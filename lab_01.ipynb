**Array**


```python
import numpy as np
list1 = [1,2,3]
array1=np.array(list1)
array1
```

```python
import numpy as np
list2=[4,5,6]
array2=np.array(list2)
array2
```

```python
type(array2)
```

```python
print("array2 is multiplied by array1:",array1*array2)
print("array2 is divided by array1:",array2/array1)
print("array2 raised to the power of array1:",array2**array1)
```

```python
print("array2 is added by array1:",array1+array2)
print("array2 is subtracted by array1:",array1-array2)
```

```python
print("sine: ",np.sin(array1))
print("natural logarithm:",np.log(array1))
print("base-10 logarithm:",np.log10(array1))
print("base-2 logarithm:",np.log2(array1))
print("Exponential:",np.exp(array1))
```

```python
people=['a','b','c','d','e']
age=[21,20,22,30,19]
weight=[55,60,50,70,45]
height=[160,135,170,168,155]
```

```python
import matplotlib.pyplot as p
```

```python
p.plot(weight,height)
p.title("relationship between height and weight of patients")
p.ylabel("height in meters")
p.xlabel("weight of patients in kgs")
```

```python
p.scatter(age,height)
p.title("relationship between height and age of patients")
p.ylabel("height in meters")
p.xlabel("weight of patients in kgs")
```

```python
p.plot(figsize=(6,4))
p.title("people weight in kgs",fontsize=16,fontstyle='italic')
# main plot bar
p.bar(x=people,height=weight,width=0.6,color='blue',edgecolor='k',alpha=0.6)
p.xlabel("people",fontsize=15)
p.xticks(fontsize=14,rotation=30)
p.yticks(fontsize=14)
p.ylabel("weight in kgs",fontsize=15)
```
